<h3 style="color:#483D8B">2.1 Identificar como se muestran los valores faltantes y estandarizarlos </h3>

```python
dfAutos.head(15)
```

```python
dfAutos.tail(10)
```

```python
dfAutos.replace("?", np.nan, inplace= True)
dfAutos.head(10)
```

<h3 style="color:#483D8B">2.2 Contar Valores faltantes por columna </h3>

```python
valoresPerdidos = dfAutos.isnull()
```

```python
valoresPerdidos.head(5)
```

```python
for column in valoresPerdidos.columns.values.tolist():
    print (column)
    print (valoresPerdidos[column].value_counts())
    print (" ")
```
